```python
import numpy as np
import healpy as hp
import os
import matplotlib.pyplot as plt
```

```python
kSZFiles = os.listdir('./Data/kSZ')
velFiles = os.listdir('./Data/velocityField')
densFiles = os.listdir('./Data/overdensity')

nside = 64
numSets=len(densFiles)
```

```python
kSZmap = np.load('./Data/kSZ/'+kSZFiles[0])
```

```python
hp.mollview(kSZmap,nest=False)
```

```python
alm = hp.sphtfunc.map2alm(kSZmap,lmax=50)
```

```python
hp.mollview(hp.alm2map(alm,nside))
```

```python
for file in kSZFiles:
    kSZmap = np.load('./Data/kSZ/'+file)
    alm = hp.sphtfunc.map2alm(kSZmap,lmax=50)
    seperatedAlm = np.transpose(np.vstack((alm.real,alm.imag)))
    np.save('./Data/kSZPS/'+file,seperatedAlm)
```

```python
for file in densFiles:
    densmap = np.load('./Data/overdensity/'+file)
    alm = hp.sphtfunc.map2alm(densmap,lmax=50)
    seperatedAlm = np.transpose(np.vstack((alm.real,alm.imag)))
    np.save('./Data/overdensityPS/'+file,seperatedAlm)
```

```python
for file in densFiles:
    velmap = np.load('./Data/velocityField/'+file)
    alm = hp.sphtfunc.map2alm(velmap,lmax=50)
    seperatedAlm = np.transpose(np.vstack((alm.real,alm.imag)))
    np.save('./Data/velocityFieldPS/'+file,seperatedAlm)
```
